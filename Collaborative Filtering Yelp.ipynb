```python
def cosine_distance(matrix, id1, id2):
    """Compute euclid distance between two rows."""    
    # only take the features that have values for both id1 and id2
    if id1 == id2:
        return 1.0
    selected_features = matrix.loc[id1].notna() & matrix.loc[id2].notna()
    
    # if no matching features, return 'not a number' (NaN)
    if not selected_features.any():
        return np.nan
    
    # get the features from the matrix
    features1 = matrix.loc[id1][selected_features]
    features2 = matrix.loc[id2][selected_features]
    
    if all(v1 == 0 for v1 in features1) or all(v2 == 0 for v2 in features2):
        return np.nan
    

    # compute the distances for the features
    distances = np.dot(features1, features2)
    features1 = np.linalg.norm(features1)
    features2 = np.linalg.norm(features2)
    

    return distances/(features1*features2)

def cosine_similarity(matrix, id1, id2):
    """Compute euclid similarity between two rows."""
    # compute distance
    distance = cosine_distance(matrix, id1, id2)
    
    # if no distance could be computed (no shared features) return a similarity of 0
    if distance is np.nan:
        return np.nan
    
    # else return similarity
    return distance

def create_similarity_matrix_cosine(matrix):
    """creates the similarity matrix based on cosine distance"""
    data = []
    for row in tqdm(matrix.index):
        row_data = []
        for col in matrix.index:
            row_data.append(cosine_similarity(matrix, col, row))
        data.append(row_data)
    similarity_matrix = pd.DataFrame(data, index=matrix.index, columns=matrix.index, dtype=float)
    return(similarity_matrix)

similarity = create_similarity_matrix_cosine(centered_utility_matrix)
display(similarity)
```

```python
def select_neighborhood(similarities, ratings, k):
    """ selects all items with similarity > 0  """
    similarities = similarities.reindex(ratings.dropna().index)
    return similarities.loc[(similarities > 0)]

def weighted_mean(neighborhood, ratings):  
    if neighborhood.empty:
        return np.nan
    else:
        return (neighborhood*ratings).sum()/neighborhood.sum()
```

```python
def number_of_businesses(ratings):
    """ determine the number of unique business id's in the data """
    return len(ratings['business_id'].unique())

def number_of_users(ratings):
    """ determine the number of unique user id's in the data """
    return len(ratings['user_id'].unique())

def number_of_ratings(ratings):
    """ count the number of ratings of a dataset """
    return ratings.shape[0]

def rating_density(ratings):
    """ compute the ratings given a data set """
    return number_of_ratings(ratings) / (number_of_businesses(ratings) * number_of_users(ratings))

def split_data(data,d = 0.75):
    """ split data in a training and test set 
       `d` is the fraction of data in the training set"""
    np.random.seed(seed=5)
    mask_test = np.random.rand(data.shape[0]) < d
    return data[mask_test], data[~mask_test]
```

```python
import random
def predict_ratings_item_based(similarity, utility, user_item_pairs):
    ratings_test_c = user_item_pairs.copy()
    for row in ratings_test_c.itertuples():
        if row.user_id in utility and row.business_id in similarity:
            neighborhood = select_neighborhood(similarity[row.business_id], utility[row.user_id], 100)
            ratings_test_c.loc[row.Index, 'predicted rating'] = weighted_mean(neighborhood, utility[row.user_id])
        else:
            ratings_test_c.loc[row.Index, 'predicted rating'] = random.uniform(0.5, 5.0)
    return ratings_test_c
```

```python
def recommended(predictions, treshold):
    return predictions[predictions['predicted rating'] >= treshold]

def hidden(predictions, treshold):
    return predictions[predictions['predicted rating'] < treshold]
```

```python
def used(predictions, treshold):
    return predictions[predictions['rating'] >= treshold]

def unused(predictions, treshold):
    return predictions[predictions['rating'] < treshold]
```

```python
def confusion(recommended, hidden, used, unused):
    TP = len(recommended.index.intersection(used.index))
    FP = len(recommended.index.intersection(unused.index))
    TN = len(hidden.index.intersection(unused.index))
    FN = len(hidden.index.intersection(used.index))
    return DataFrame([[TP, FP], [FN, TN]], ['recommended', 'hidden'], ['used', 'unused'])
```

```python
def precision(confusion_matrix):
    TP = confusion_matrix.loc['recommended', 'used']
    FP = confusion_matrix.loc['recommended', 'unused']
    return TP/(TP+FP)

def recall(confusion_matrix):
    TP = confusion_matrix.loc['recommended', 'used']
    FP = confusion_matrix.loc['hidden', 'used']
    return TP/(TP+FP)

def F1_score(precision, recall):
    return 2*(precision*recall)/(precision+recall)
```
